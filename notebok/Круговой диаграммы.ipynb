# Круговой диаграммы

```python
 import matplotlib.pyplot as plt 
 
 x = [15, 25, 25, 30, 5] 
 labels = ['Ваньянц И', 'Бокань И', 'Симоненко А', 'Егор В', 'Чернова С'] 
 colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red'] 
 
 fig, ax = plt.subplots() 
 ax.pie(x, labels = labels, colors = colors) 
 ax.set_title('Выборов СКФУ') 
 plt.show() 
```

```python
 import matplotlib.pyplot as plt 
 
 x = [15, 25, 25, 30, 5] 
 labels = ['Ваньянц И', 'Бокань И', 'Симоненко А', 'Егор В', 'Чернова С'] 
 colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red'] 
 explode = [0, 0, 0, 0, 0.2] 
 
 fig, ax = plt.subplots() 
 ax.pie(x, labels = labels, colors = colors, autopct='%.0f%%', explode = explode) 
 ax.set_title('Выборов СКФУ') 
 plt.show() 
```

```python
 import matplotlib.pyplot as plt 
 
 x = [15, 25, 25, 30, 5] 
 labels = ['Ваньянц И', 'Бокань И', 'Симоненко А', 'Егор В', 'Чернова С'] 
 colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red'] 
 explode = [0, 0, 0, 0, 0.2] 
 
 fig, ax = plt.subplots() 
 ax.pie(x, labels = labels, 
 colors = colors, 
 autopct='%.0f%%', 
 explode = explode, 
 shadow = True, 
 startangle = 180) 
 
 ax.set_title('Выборов СКФУ')
 plt.show() 
```
